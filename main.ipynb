# Parte 1: El Problema del Bandido de k-brazos

Esta sección presenta el problema del bandido de $k$-brazos, un escenario clásico de aprendizaje por refuerzo que ilustra el dilema exploración-explotación. Para su estudio, se ha implementado un entorno simulado donde un agente interactúa con $k$ opciones, evaluando tres distribuciones de recompensa distintas: Bernoulli, Binomial y Normal.

## Parte 2: Aprendizaje en entornos complejos

Esta sección presenta la extensión del aprendizaje por refuerzo hacia entornos complejos, modelados como Procesos de Decisión de Markov (MDP), donde las acciones del agente tienen consecuencias secuenciales a largo plazo. Para su estudio, se ha implementado una progresión de simulaciones que transita desde laberintos discretos generados procedimentalmente hasta entornos dinámicos de alta dimensionalidad (Flappy Bird), evaluando la evolución desde métodos tabulares clásicos (SARSA, Q-Learning, Monte Carlo) hacia arquitecturas avanzadas de aproximación de funciones y Deep Reinforcement Learning (DQN).
