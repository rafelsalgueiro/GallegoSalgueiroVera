## 1. Introducción

El aprendizaje por refuerzo es una rama de la inteligencia artificial centrada en cómo un agente debe interactuar con su entorno para aprender a tomar decisiones óptimas mediante la experiencia y la retroalimentación. El problema del bandido multibrazo (k-armed bandit) representa el ejemplo más puro del dilema de la exploración-explotación dentro de este campo, ya que modela situaciones de toma de decisiones secuenciales bajo incertidumbre en entornos estacionarios. En este escenario, el agente se enfrenta a una máquina con k brazos o acciones, donde cada uno posee una distribución de recompensa desconocida Pi que debe ser descubierta mediante la interacción. 

El dilema central reside en encontrar el equilibrio adecuado entre dos estrategias opuestas: la explotación de los brazos que, según la experiencia previa, han mostrado mejores resultados, y la exploración de opciones menos conocidas para identificar si alguna posee una recompensa esperada superior. Para medir el éxito de estas estrategias, no solo se analiza la recompensa acumulada, sino también el rechazo. El rechazo cuantifica la pérdida total que sufre el agente por no elegir siempre el brazo óptimo, siendo el objetivo de los mejores algoritmos minimizar este valor a lo largo de un horizonte temporal determinado.

## 2. Algoritmos Implementados

### 2.1 Epsilon-Greedy 
El algoritmo epsilon-greedy es una estrategia simple pero efectiva para abordar el dilema de la exploración-explotación. En cada paso, el agente elige la acción con la mayor recompensa esperada con una probabilidad de 1 - ε, y con una probabilidad de ε, selecciona una acción al azar. Este enfoque permite al agente explorar nuevas acciones mientras sigue explotando las que han demostrado ser rentables. Sin embargo, la elección del valor de ε es crucial, ya que un valor demasiado alto puede llevar a una exploración excesiva, mientras que un valor demasiado bajo puede resultar en una explotación prematura.

### 2.2 Epsilon-Greedy con Decay
El algoritmo epsilon-greedy con decay introduce una variación al método estándar al reducir gradualmente el valor de ε a medida que avanza el tiempo. Esto permite al agente explorar más al principio, cuando tiene poca información sobre las recompensas de los brazos, y luego enfocarse en la explotación a medida que adquiere más conocimiento. La tasa de decaimiento de ε es un parámetro importante que debe ser ajustado para asegurar un equilibrio adecuado entre exploración y explotación a lo largo del tiempo.


### 2.3 UCB1 
El algoritmo Upper Confidence Bound (UCB1) es una estrategia que selecciona la acción con el mayor valor de recompensa esperada más un término de exploración que disminuye a medida que se selecciona esa acción. Este enfoque garantiza que cada acción sea explorada un número suficiente de veces, lo que permite al agente identificar el brazo óptimo de manera eficiente. UCB1 es especialmente efectivo en entornos donde las recompensas son ruidosas, ya que su término de exploración ayuda a mitigar la incertidumbre.

### 2.4 Softmax
El algoritmo Softmax es una estrategia de selección de acciones que asigna probabilidades a cada acción en función de sus recompensas esperadas. En lugar de elegir la acción con la mayor recompensa esperada, el agente selecciona acciones de manera probabilística, lo que permite una exploración más equilibrada. La temperatura del Softmax es un parámetro que controla el grado de exploración: valores altos conducen a una exploración más uniforme, mientras que valores bajos favorecen la explotación de las acciones con mayores recompensas esperadas.

## 4. Conclusiones

#### Epsilon-Greedy con decaimiento
Es el algoritmo más polivalente y consistente a lo largo de todo el estudio. Mantiene un crecimiento constante y estable sin importar si interactúa con brazos de tipo Bernoulli, Binomial o Normal. Su lógica basada puramente en promedios empíricos lo convierte en la elección más segura cuando no se tiene información previa sobre el tipo de distribución del entorno.

#### UCB1 
Su rendimiento es altamente dependiente de la varianza del problema.

- Óptimo para brazos Normales (alta dificultad): Es el ganador absoluto en entornos con mucho ruido. Su cota superior de confianza está diseñada precisamente para no dejarse engañar por las grandes fluctuaciones de una distribución Normal compleja.

- Poco eficiente en brazos Binomiales/Bernoulli: Aunque logra identificar el mejor brazo casi siempre (alcanzando un 96% de selecciones óptimas en el bandido Binomial), su exceso de precaución le obliga a explorar demasiado en escenarios que son simples y predecibles, perdiendo eficiencia en los primeros pasos.

#### Softmax 
Es el algoritmo que peor soporta el cambio entre bandidos, ya que su fórmula exponencial es extremadamente sensible a la escala numérica de las recompensas.

- Requiere una calibración exacta para cada tipo de brazo. Por ejemplo, una temperatura que funciona bien para recompensas pequeñas (Bernoulli) falla de forma drástica en el bandido Binomial (cayendo a un 41.4% de tasa de acierto) porque la escala de los números cambia por completo. Es una opción de alto riesgo si las condiciones del entorno varían.
