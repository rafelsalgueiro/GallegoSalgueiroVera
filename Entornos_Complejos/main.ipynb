## 2. Algoritmos Implementados

### 2.1 Métodos Tabulares (MC y TD)

Los algoritmos tabulares almacenan el valor esperado de cada par estado-acción en una tabla (matriz Q).

- Monte Carlo (On-Policy / Off-Policy): Aprende de episodios completos, actualizando las estimaciones basándose en los retornos reales observados al final de cada trayectoria. Es insesgado pero sufre de alta varianza.

- SARSA (On-Policy TD): Algoritmo de Diferencia Temporal que actualiza los valores paso a paso, tomando en cuenta la política actual del agente (incluyendo su exploración). Tiende a encontrar rutas más seguras.

- Q-Learning (Off-Policy TD): Actualiza sus estimaciones asumiendo que siempre se tomará la mejor acción futura, independientemente de la política de exploración que esté siguiendo. Es más agresivo a la hora de buscar la ruta óptima.

### 2.2 Semi-Gradient SARSA

Cuando el espacio de estados es continuo o demasiado grande (ej. coordenadas, velocidades), es imposible mantener una tabla. Semi-Gradient SARSA resuelve esto utilizando Aproximación de Funciones. Representa la función de valor Q mediante una combinación lineal de características (pesos) y actualiza estos pesos utilizando el gradiente del error de diferencia temporal. Es el puente clásico hacia el Deep RL.

### 2.3 Deep Q-Network (DQN)

DQN revoluciona el aprendizaje por refuerzo al utilizar una Red Neuronal Artificial como aproximador de la función Q.  Para superar la inestabilidad inherente de entrenar redes neuronales con datos secuenciales correlacionados, implementa dos mecanismos críticos: un Replay Buffer (memoria de experiencias pasadas de la que se toman muestras aleatorias) y una Red Objetivo (Target Network) que estabiliza las actualizaciones.
